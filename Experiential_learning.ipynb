```python
# Creating the Appointment class
class Appointment:
    def __init__(self, Patient_name, Doctor_name, Appointment_data):
        self.__Patient_name = Patient_name
        self.__Doctor_name = Doctor_name
        self.__Appointment_data = Appointment_data
        
    def display(self):
        return f"Patient_name: {self.__Patient_name}, doctor_name: {self.__Doctor_name}, Appointment: {self.__Appointment_data}"
# Creating the Specialized_appointment  
class Specialized_appointment(Appointment):
    def __init__(self, Patient_name, Doctor_name, Appointment_data, Specialized_appointment):
        super().__init__(Patient_name,Doctor_name, Appointment_data)
        self.__Specialized_appointment = Specialized_appointment
        
    def display(self):
        return f"Patient_name: {self.__Patient_name}, doctor_name: {self.__Doctor_name}, Appointment: {self.__Appointment_data}, Specialized_appointment: {self.__Specialized_appointment}"
# Creating the emergency_appointment
class Emergency_appointment(Appointment):
    def __init__(self, Patient_name, Doctor_name, Appointment_data, priority):
        super().__init__(Patient_name,Doctor_name, Appointment_data)
        self.__priority = priority
        
    def display(self):
        return f"Patient_name: {self.__Patient_name}, doctor_name: {self.__Doctor_name}, Appointment: {self.__Appointment_data}, priority: {self.__priority}"
# Creating the LabTest class 
class LabTest:
    def __init__(self, patient_name, test_name, test_date):
        self.__patient_name = patient_name
        self.__test_name = test_name
        self.__test_date = test_date

    def dis(self):
        return f"Patient_name: {self.__patient_name}, test_name: {self.__test_name}, test_date: {self.__test_date}"

class Hospital_management_system:
    def __init__(self):
        self.__appointments = [] # Creating the empty list for appending the values
        self.__lab_tests = [] # same here

    def add_appointment(self, appointment):
        self.__appointments.append(appointment)
        print("Appointment added successfully.") # Successfully added

    def add_lab_test(self, lab_test):
        self.__lab_tests.append(lab_test)
        print("Lab test added successfully.") # Successfully added

    # Creating the getter
    def get_appointment(self):
        return self.__appointments

    def get_lab_test(self):
        return self.__lab_tests

    # showing all the appointments
    def show_all_appointments(self):
        if not self.__appointments:
            print("No appointments available.")
        else:
            print("All Appointments:")
            for appointment in self.__appointments:
                print(appointment)
    # Showing all lab test
    def show_all_lab_tests(self):
        if not self.__lab_tests:
            print("No lab tests available.")
        else:
            print("All Lab Tests:")
            for lab_test in self.__lab_tests:
                print(lab_test)

# Creating the Object
hms = Hospital_management_system()

# adding the appointments
hms.add_appointment(Appointment("Harry", "Dr. Smith", "2024-12-21"))
hms.add_appointment(Specialized_appointment("Ron", "Dr. Brown", "2024-12-22", "Cardiology"))
hms.add_appointment(Specialized_appointment("Draco", "Dr. White", "2024-12-23", "High"))

# Adding the lab_tests
hms.add_lab_test(LabTest("Harry", "Blood Test", "2024-12-24"))
hms.add_lab_test(LabTest("Ron", "MRI Scan", "2024-12-25"))

# displaying the appointments and the tests
print(hms.show_all_appointments())
print(hms.show_all_lab_tests())
    

        
```

```python
class ECommerceSystem:
    platform_tax_rate = 0.1  # Default tax rate = 10%

    @staticmethod # Applying the static method
    def calculate_tax(price, tax_rate):
        return price * tax_rate

    @classmethod # Applying the class method
    def update_tax_rate(cls, new_rate):
        cls.platform_tax_rate = new_rate
        print(f"Platform tax rate updated to {new_rate * 100:.1f}%")

    def __init__(self, base_price):
        self.base_price = base_price

    def calculate_total(self):
        tax = ECommerceSystem.calculate_tax(self.base_price, ECommerceSystem.platform_tax_rate)
        total_price = self.base_price + tax
        return total_price


# Default tax rate
order1 = ECommerceSystem(100)
print(f"Order1 Total including default tax: {order1.calculate_total()}")

# Update the platform tax rate
ECommerceSystem.update_tax_rate(0.15)  # Updating tax rate to 15%

# New orders with updated tax rate
order2 = ECommerceSystem(200)
print(f"Order2 Total with upgraded tax: {order2.calculate_total()}")
```

```python
class BankAccount:
    minimum_balance = 1000  # Global minimum balance for all accounts

    def __init__(self, holder_name, balance):
        self.holder_name = holder_name
        self.balance = balance

    @classmethod
    def update_minimum_balance(cls, new_minimum):
        
        cls.minimum_balance = new_minimum
        print(f"Minimum balance updated to {cls.minimum_balance}")

    @staticmethod
    def calculate_interest(balance, interest_rate):
        
        return balance * (interest_rate / 100)

    def deposit(self, amount):
        if amount <= 0:
            print("Please Enter Deposit value in positive numbers")
        else:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif self.balance - amount < BankAccount.minimum_balance:
            print(f"Cannot withdraw {amount}. Minimum balance of {BankAccount.minimum_balance} must be maintained.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def display_account_info(self):
        print(f"Account Holder: {self.holder_name}, Balance: {self.balance}")


# Example usage
# Create accounts
account1 = BankAccount("Harry Potter", 5000)
account2 = BankAccount("Hermione Granger", 3000)

# Display account info
account1.display_account_info()
account2.display_account_info()

# Deposit and withdraw operations
account1.deposit(2000)
account1.withdraw(7000)  # Failing due to minimum balance
account1.withdraw(2000)

# Calculate interest
interest = BankAccount.calculate_interest(account1.balance, 5)
print(f"Interest on {account1.balance} at 5%: {interest}")

# Update minimum balance
BankAccount.update_minimum_balance(1500)
account2.withdraw(2000)  # Again failing due to minimum balance


# Display final account info
account1.display_account_info()
account2.display_account_info()
```
